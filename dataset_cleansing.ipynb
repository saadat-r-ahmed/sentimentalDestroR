## SemEval17 Task 4

```python
import pandas as pd
semEvalRawPath = 'Sentiment_Dataset/SemEval17/Raw/SemEval2017-task4-dev.subtask-A.english.INPUT.txt'
data = pd.read_csv(semEvalRawPath, names  = ['id', 'polarity', 'text', 'date'],sep = '\t')
```

```python
data.drop(['id', 'date'], axis=1, inplace = True)
```

```python
data
```

```python
data['polarity'].replace('neutral', 'NEU', inplace = True)
data['polarity'].replace('negative', 'NEG', inplace = True)
data['polarity'].replace('positive', 'POS', inplace = True)
```

```python
data
```

```python
!mkdir ./Sentiment_Dataset/SemEval17/Clean
data.to_csv("./Sentiment_Dataset/SemEval17/Clean/semeval.csv", encoding = 'utf-8', index=False)
```
